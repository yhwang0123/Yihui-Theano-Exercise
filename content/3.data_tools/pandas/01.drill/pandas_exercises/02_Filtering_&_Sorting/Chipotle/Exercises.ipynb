### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url= "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url,'\t')
chipo
```

### Step 4. How many products cost more than $10.00?

```python
# clean the item_price column and transform it in a float
prices= [float(value[1 : -1]) for value in chipo.item_price]

# reassign the column with the cleaned prices
chipo.item_price = prices

# delete the duplicates in item_name and quantity
chipo_filtered = chipo.drop_duplicates(subset=['item_name','quantity'])


# select only the products with quantity equals to 1
chipo_filtered_quantity=chipo_filtered[(chipo_filtered["quantity"]==1)]

#select item price more than 10
chipo_filted_price=chipo_filtered_quantity[chipo_filtered_quantity["item_price"]>10]

print(chipo_filted_price.quantity.sum())
```

```python
chipo_filted_price
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
# delete the duplicates in item_name and quantity
chipo_filtered = chipo.drop_duplicates(subset=['item_name','quantity'])

# select only the products with quantity equals to 1
chipo_filtered_quantity=chipo_filtered[(chipo_filtered["quantity"]==1)]

# select only the item_name and item_price columns
chipo_twocolumns=chipo_filtered_quantity[['item_name', 'item_price']]

# sort the values from the most to less expensive
chipo_twocolumns.sort_values(by=['item_price'],ascending=False)
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by=['item_name'])
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.iloc[[chipo['item_price'].idxmax()]]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python

chipo[chipo.item_name == "Veggie Salad Bowl"].agg({'quantity': 'sum'})
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo_multi_soda=chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)]
print(len(chipo_multi_soda))
```
